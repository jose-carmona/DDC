```python
df = pd.read_csv('../data/metadata_Xception.csv')
```

```python
df.sample(10)
```

```python
for index, row in df.query('label == "FAKE" and dir != "dfdc_train_part_1"').sample(50).iterrows():
    print(f'file {row.video} // {row.original}')

    vc1 = VideoFileClip(f"../videos/{row.dir}/{row.video}")
    audioclip1 = vc1.audio
    vc2 = VideoFileClip(f"../videos/{row.dir}/{row.original}")
    audioclip2 = vc2.audio
    if np.array_equal(audioclip1.to_soundarray(), audioclip2.to_soundarray()):
        print(f'{row.video} == {row.original}')
    else:
        print(f'{row.video} != {row.original}')
```
